```python
import math
import numpy as np
import matplotlib.pyplot as plt
 
img = np.ones((1000,1000,3))
 
point1 = (300,100)
point2 = (100,100)
point3 = (100,300)
point4 = (300,300)
 
```

```python
def MoveObject(x1, y1, x2, y2, x3, y3, x4, y4):
   
    move_x = 200
    move_y = 200
    
    x1 = x1 + move_x
    y1 = y1 + move_y
    x2 = x2 + move_x
    y2 = y2 + move_y
    x3 = x3 + move_x
    y3 = y3 + move_y
    x4 = x4 + move_x
    y4 = y4 + move_y
    
    x5 = [x1,x2,x3,x4,x1]
    y5 = [y1,y2,y3,y4,y1]
    plt.plot(x5,y5)
    
def ScaleObject(x1, y1, x2, y2, x3, y3, x4, y4):
    scale_x = 3
    scale_y = 3
    
    x1 = x1 * scale_x
    y1 = y1 * scale_y
    x2 = x2 * scale_x
    y2 = y2 * scale_y
    x3 = x3 * scale_x
    y3 = y3 * scale_y
    x4 = x4 * scale_x
    y4 = y4 * scale_y
    
    x5 = [x1,x2,x3,x4,x1]
    y5 = [y1,y2,y3,y4,y1]
    plt.plot(x5,y5)
    
def RotateObject(x1, y1, x2, y2, x3, y3, x4, y4, angle):
    
    xconst = point1[0]
    yconst = point1[1]
    x1 = (xconst-200) * math.cos(angle) - (yconst-200)*math.sin(angle) + 200
    xconst = point1[0]
    yconst = point1[1]
    y1 = (xconst-200) * math.sin(angle) + (yconst-200)*math.cos(angle) + 200
    xconst = point2[0]
    yconst = point2[1]
    x2 = (xconst-200) * math.cos(angle) - (yconst-200)*math.sin(angle) + 200
    xconst = point2[0]
    yconst=point2[1]
    y2 = (xconst-200) * math.sin(angle) + (yconst-200)*math.cos(angle) + 200
    xconst=point3[0]
    yconst=point3[1]
    x3 = (xconst-200) * math.cos(angle) - (yconst-200)*math.sin(angle) + 200
    xconst=point3[0]
    yconst=point3[1]
    y3 = (xconst-200) * math.sin(angle) + (yconst-200)*math.cos(angle) + 200
    xconst=point4[0]
    yconst=point4[1]
    x4 = (xconst-200) * math.cos(angle) - (yconst-200)*math.sin(angle) + 200
    xconst=point4[0]
    yconst=point4[1]
    y4 = (xconst-200) * math.sin(angle) + (yconst-200)*math.cos(angle) + 200
    x5 = [x1,x2,x3,x4,x1]
    y5 = [y1,y2,y3,y4,y1]
    plt.plot(x5,y5)
  
```
